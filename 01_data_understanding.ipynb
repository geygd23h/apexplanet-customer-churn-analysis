# Exploratory Data Analysis

Objective:
Identify factors associated with customer churn and uncover actionable business insights.

# Key Findings

1. Month-to-Month customers have the highest churn rate (45.84%), while Two-Year contract customers have the lowest churn rate (2.55%)

2. Customers who churn pay higher monthly charges ($74.44) compared to retained customers ($61.27).

3. Churned customers have significantly lower average tenure (17.98 months) than retained customers (37.59 months).

4. Approximately 26.54% of customers have churned, indicating a meaningful customer retention challenge.

5. Missing values in Internet Type are not data quality issues; they correspond to customers without Internet Service.

```python
df[['Offer','Internet Type','Churn Category','Churn Reason']].isnull().sum()
```

```python
df['Offer'].value_counts(dropna=False)
```

```python
df['Customer ID'].nunique()
```

```python
df['Offer'] = df['Offer'].fillna('No Offer')
```

```python
df['Churn Category'] = df['Churn Category'].fillna('Not Churned')

df['Churn Reason'] = df['Churn Reason'].fillna('Not Churned')
```

# Final Business Insights

1. Approximately 26.54% of customers have churned.

2. Month-to-Month customers have the highest churn rate (45.84%), while Two-Year contract customers have the lowest churn rate (2.55%).

3. Churned customers pay higher average monthly charges ($74.44) than retained customers ($61.27).

4. Churned customers have significantly lower tenure (17.98 months) than retained customers (37.59 months).

5. Customer Satisfaction Score shows the strongest negative correlation with churn (-0.495), making it a key retention indicator.

6. Customers with more referrals and longer tenure are less likely to churn.

7. Missing values in Internet Type, Churn Category, and Churn Reason were business-related and not data quality errors.
